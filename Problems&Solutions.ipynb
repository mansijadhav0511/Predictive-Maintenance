```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import numpy as np
from scipy.stats import zscore

# Load the dataset
file_path = 'C:/Users/MEGHNA/Desktop/Project 1/predictive-maintenance-dataset.csv'
data = pd.read_csv(file_path)

# Handling Missing Values
imputer = SimpleImputer(strategy='mean')  # Using mean imputation for example
data['vibration'] = imputer.fit_transform(data[['vibration']])

# Checking Data Types
data.info()

# Outlier Detection and Removal using Z-score
data = data[(np.abs(zscore(data.select_dtypes(include=['float64']))) < 3).all(axis=1)]

# Normalization/Standardization
scaler = StandardScaler()
data[data.columns[1:]] = scaler.fit_transform(data[data.columns[1:]])

# Save the cleaned data
cleaned_file_path = 'C:/Users/MEGHNA/Desktop/Project 1/cleaned_predictive_maintenance_dataset.csv'
data.to_csv(cleaned_file_path, index=False)

# Display the first few rows of the cleaned dataset
data.head()
```

```python
Loading the dataset: Reads the CSV file from the specified path.
Handling Missing Values: Fills missing values in the 'vibration' column using mean imputation.
Checking Data Types: Displays the data types and non-null count for each column.
Outlier Detection and Removal: Uses Z-score to identify and remove outliers from float columns.
Normalization/Standardization: Scales all features to have a mean of 0 and a standard deviation of 1.
Save the cleaned data: Writes the cleaned dataset back to a CSV file
```
